# Visual Data Analysis of Fraudulent Transactions

```python
# initial imports
import pandas as pd
import calendar
import plotly.express as px
import hvplot.pandas
from sqlalchemy import create_engine
```

```python
# create a connection to the database
engine = create_engine("postgresql://postgres:postgres@localhost:5432/suspicious_transactions_db")
```

```python
# loading data for transactions from the database
query1 = "SELECT id_card_holder, date, amount FROM credit_card JOIN  card_holder ON id_card_holder=id JOIN transaction ON credit_card.card=transaction.card AND date_part('hour', date) BETWEEN 7 AND 9 ;"

transaction_7_9 = pd.read_sql(query1, engine)
transaction_7_9.sort_values("amount", ascending=False, inplace=True)
highest_100_transactions = transaction_7_9.iloc[:100,:]

highest_100_transactions.reset_index()
```

```python
 # loading data for card holder 2 and 18 from the database
query2 = "SELECT id_card_holder, date, amount FROM credit_card JOIN  card_holder ON id_card_holder=id JOIN transaction ON credit_card.card=transaction.card WHERE card_holder.id IN (2,18);"



transaction_df = pd.read_sql(query2, engine)

transaction_df
```

```python
# plot for cardholder 2
id_2= transaction_df.loc[transaction_df["id_card_holder"]==2]
credit_card_2 = id_2.hvplot.line( x="date", y="amount")
credit_card_2 
```

```python
# plot for cardholder 18
id_18= transaction_df.loc[transaction_df["id_card_holder"]==18]
credit_card_18 = id_18.hvplot.line( x="date", y="amount")
credit_card_18 
```

```python
 # combined plot for card holders 2 and 18
combined_plot = credit_card_2*credit_card_18.opts(logy=True)
combined_plot
```
